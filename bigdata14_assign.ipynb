```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
```

과제1) 아래 미국 공항 traffic 데이터를 이용하여 미국 공항별 fraffic을 지도에 나타내시오.   
marker의 color는 cnt를 기준으로 선택되게.     
marker의 size는 고정.  
 colormap 등 다른 option은 예제와 다르게 변경.   
 지도에 미국만 나타나도록 변경  

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2011_february_us_airport_traffic.csv')

df['text'] = df['airport'] + ', ' + df['city'] + ', ' + df['state'] + ' - Arrivals: ' + df['cnt'].astype(str)

fig = px.scatter_geo(df,
                      lat='lat', lon='long',
                      color='cnt',
                      size_max=15,  # Marker 크기 고정
                      projection='natural earth',
                      color_continuous_scale='Viridis',  # Viridis colormap 사용
                      title='미국 공항별 트래픽',
                      labels={'cnt': '도착 횟수'},
                      )

fig.update_geos(showcountries=True, countrycolor="black")

fig.update_layout(geo=dict(bgcolor='#fafafa'))

fig.show()
```

과제2) 아래의 gapminder 데이터 ( 아프리카) 를 이용하여 scatter geo figure를 작성하시오.  
국가 코드를 이용하여 지도 작성.      
-marker의 크기가 'pop' 에 의해 결정   
-marker의 color는 "country"에 의해 결정.  
-animation은 year 기준으로.   
-지도에 아프리카만 나타나도록
'  


```python
import plotly.express as px
import pandas as pd

df = px.data.gapminder().query("continent=='Africa' ")

fig = px.scatter_geo(df,
                      locations='iso_alpha',
                      size='pop',
                      color='country',
                      animation_frame='year',
                      projection='natural earth',
                      title='아프리카 국가별 인구 변화',
                      labels={'pop': '인구', 'country': '국가'},
                      )
fig.update_layout(geo=dict(bgcolor='#fafafa'))

fig.show()

```

과제3) 아래의 2007년 유럽의 gapminder 데이터를 이용하여 chropleth map을 작성하시오.
 - 표현 값은 'gdpPercap'
 - 지도에 europe 만 나오게

```python
df_a = px.data.gapminder().query( "continent=='Europe' " )
df_g=df_a.query("year==2007")
fig = px.choropleth(df_g,
                    locations='iso_alpha',
                    color='gdpPercap',
                    hover_name='country',
                    projection='natural earth',
                    title='2007년 유럽 국가별 GDP per Capita',
                    labels={'gdpPercap': 'GDP per Capita', 'country': '국가'},
                    )

fig.update_layout(geo=dict(bgcolor='#fafafa'))
fig.show()
```
